```python
# making a copy of Train & Test 
Train.copy(),
Test.copy()
```

```python
# dropping some columns
X = Train.drop(["ID","Target"], axis=1)
y = Train["Target"]
X_test = Test.drop(["ID"], axis=1)
```

```python
#ENCODING OF THE CATEGORICAL FEATURES
from sklearn.preprocessing import LabelEncoder
```

```python
for i in ["country","urban_or_rural"]:
    X[i]= LabelEncoder().fit_transform(X[i])
    X_test[i]=LabelEncoder().fit_transform(X_test[i])
```

```python
X.head()
```

```python
X_test.head()
```

```python
# SPILTING DATA INTO TRAIN & VALIDATION SETS
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state=1000)
```

```python
# MODEL BUILDING PART2
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import log_loss, accuracy_score
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([("scaler", StandardScaler()),("model", LinearRegression())])
```

```python
pipe.fit(X_train,y_train)
pipe.score(X_val, y_val)
```

```python
pred= pipe.predict(X_val)
pred
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kfolds = KFold(n_splits=10, random_state= 1000, shuffle= True)
model = LinearRegression()
scoring = 'neg_mean_absolute_error'
results = cross_val_score(model, X, y, cv=kfolds, scoring=scoring)
results.mean()
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_val)
prediction
```

```python
# making prediction on unseen data (X_test)
#fitting the entire data

model.fit(X, y)

pred = model.predict(X_test)
pred
```

# SPOT-CHECKING OF ALGORITHMS

```python
from sklearn.linear_model import Ridge
```

```python
model2 = Ridge()
scoring = 'neg_mean_absolute_error'
results = cross_val_score(model2, X, y, cv=kfolds, scoring=scoring)
results.mean()
```
